### Sentiment for Dutch statements

This notebook requires Python 2 (2.5+), because it uses the [pattern](https://www.clips.uantwerpen.be/pages/pattern-nl) package. This is a text mining package for Dutch which supports sentiment analysis.

```python
from pattern.nl import sentiment
import pandas as pd
import json

settings_file = 'D:/thesis/settings - nl.json'
```

```python
#Preparation

#Read settings
settings = json.loads(open(settings_file).read())["settings"]

#Read data
df = pd.read_json(settings['mentions_json'], compression = 'gzip')
df.sort_index(inplace = True)
```

```python
res = df['TEXT'].apply(lambda x: sentiment(x))
df['POLARITY'] = res.apply(lambda x: x[0])
df['SUBJECTIVITY'] = res.apply(lambda x: x[1])
del res
```
